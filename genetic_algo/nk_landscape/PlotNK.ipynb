```python
if __name__=='__main__':
    N=16; K=10;
    # Keep N < 20 in order to run it in less than 10 seconds.
    m16_4 = NK(N, K)
    x = GrayCode(N)
    y = np.zeros((2**N))
    for i, genome in enumerate(x):
        y[i] = m16_4.get_value(genome)
#     codes=GrayCode(N)
#     print('%d digits gray codes:' % N)
#     print(codes)
#    print('Using Iterator:')
    #for c in GrayCode(20):  
    #    print c
#     for c in GrayCodeIterator(N):
#         print(c)
```

# Plotting

```python
Z = y.reshape( (2**(floor(N/2)), 2**(ceil(N/2))) )
fig,ax =plt.subplots(1)
ax.plot(y)
ax.set_xlabel('genome (in grey code)')
ax.set_ylabel('fitness')
#plt.imshow(Z)
```

```python
plt.imshow(Z)
```

```python
X = np.arange(2**(floor(N/2)))
Y = np.arange(2**(ceil(N/2)))
X, Y = np.meshgrid(X, Y)

# Copied
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
fig = plt.figure(figsize=(25,18))
ax = fig.gca(projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-0.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
